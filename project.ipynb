```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']


'''
comment
'''
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
import matplotlib.pyplot as plt

# data for the bar plot
algorithms = ['Actual Score', 'XGBoost', 'Random Forest', 'Linear Regression', 'Catboost', 'LightGBM', 'Lasso Regression']
matches = ['BAN vs PAK', 'SL vs AFG', 'IND vs SA', 'WI vs AUS', 'ENG vs NZ']
scores = [[127, 105, 211, 157, 191],  # real
          [141, 142, 177, 142, 183],  # xgboost
          [149, 146, 181, 152, 182],  # random forest
          [128, 133, 127, 131, 129],  # linear
          [128, 148, 168, 136, 187],  # catboost
          [131,110 , 156, 124, 140],  # lightgbm
          [143, 137, 179, 146, 171]]  # lasso regression

# set the bar plot parameters
bar_width = 0.1
opacity = 0.8
colors = ['#00FFFF', '#ff7f0e', 'g', 'r', '#AAA7A6', '#e5ae38', 'b']

# create the figure and axis objects
fig, ax = plt.subplots(figsize=(8, 6))

# plot the bars
for i in range(len(algorithms)):
    x = [j + (i * bar_width) for j in range(len(matches))]
    ax.bar(x, scores[i], color=colors[i], width=bar_width, edgecolor='white', label=algorithms[i], alpha=opacity)

# add some text for labels, title and legend
ax.set_xlabel('Matches')
ax.set_ylabel('Scores')
ax.set_title('Scores predicted after completion of 8 overs')
ax.set_xticks([j + (len(algorithms)/2) * bar_width for j in range(len(matches))])
ax.set_xticklabels(matches)

# adjust the legend position
ax.legend(bbox_to_anchor=(1.04,1), loc="upper left")

# show the bar plot
plt.show()
```
